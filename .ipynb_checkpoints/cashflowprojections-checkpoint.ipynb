```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('parking.csv')


convert_columns = list(df.columns)[1:4]
for col in convert_columns: 
    def convert(row):
        row = row.replace(',', "")
        row = int(row)
        return row
    df[col] = df[col].apply(convert)

df = df.drop(columns=['Total Available Parking', 'Total off Street Parking'])

price_columns = list(df.columns)[2:]
for col in price_columns: 
    def convert(row):
        row = row.replace('$', "")
        row = float(row)
        return row
    df[col] = df[col].apply(convert)

for i in range(len(df)):
    city = df['City '][i]
    city = city.replace('Metro', "")
    df['City '][i] = city
    
df.columns = ['City', 'Total', 'Hourly', "Daily", 'Monthly']

df['Hourly Total'] = df['Total'] * .2 * df['Hourly']
df['Daily Total'] = df['Total'] * .3 * df['Daily']
df['Monthly Total'] = df['Total'] * .5 * df['Monthly']
df['Yearly Revenue'] = (df['Hourly Total'] + df['Daily Total'] + df['Monthly Total']) * .3
```

```python
market_share = np.array([.001, .01, .08,.15,.20])

def make_bar(market_share, case):
    fig, ax = plt.subplots(figsize=(10,10))
    width = 0.35
    index = np.arange(5)
    market_share = np.array(market_share)
    ax.bar(index, df['Yearly Revenue'].sum() * market_share, width, color='orange', zorder=3)
    ax.bar(index + width, 3000000 * np.array([i for i in range(1, 6)]) + ((df['Yearly Revenue'].sum() * market_share) * .3), width, color='#5683f5', zorder=3)
    ax.set_facecolor('black')

    ax.set_xticks((index+width))
    ax.set_xticklabels([f"Year {i}" for i in range(1,6)])
    ax.grid(color='white', linestyle='dotted', alpha=0.5, zorder=-1)
    ax.set_title('Cash Flow Projections', fontsize=20)
    ax.set_xlabel('Years in Operation', fontsize=15)
    ax.set_ylabel('$ Millions (10s)', fontsize=15)

    for x, y in zip(index, market_share):
        total = df['Yearly Revenue'].sum()
        cost = 3000000 * (x + 1) + ((total * y) * .1)
        ax.text(x, total * y + 1000000, "${:,.2f}".format(round(total * y, 2)), horizontalalignment='center', color='white', fontsize=12)
    
    plt.savefig(f"Cash Flow Projections - {case}.png")
        
market = pd.DataFrame(columns=[f'Year {i}' for i in range(1,6)])
for x, y in zip([1, .8, .6], ['Best', 'Average', 'Worst']):
    temp = [i * x for i in market_share]
    make_bar(temp, y)
    temp_df = pd.DataFrame(temp).T
    temp_df.columns = [f'Year {i}' for i in range(1,6)]
    market = pd.concat([market, temp_df])
```

```python
market.index = ['Best', 'Average', 'Worst']
market.to_csv('margin.csv')
```
